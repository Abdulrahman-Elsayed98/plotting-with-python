```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.title("Gas Prices Over Time ($)")
plt.xlabel('Year')
plt.ylabel('Price ($)')

plt.xticks(gas.Year[::2])

plt.plot(gas.Year, gas.USA, label = 'USA')
plt.plot(gas.Year, gas.Canada, label = 'Canada')
plt.plot(gas.Year, gas['South Korea'], label = 'South Korea')

plt.legend()
plt.grid(True)

plt.savefig('Gas price over time.png', dpi = 250)
plt.show()
```
